## コラム（「10進数 ⇔ 2進数」できるかな？）

**Q1** 2進数の 10101 は 10進数で何？

- 10101 は 2進数で5桁の数．
- 10進数に1の位，10の位，100の位があるように，2進数には1の位，2の位，4の位，8の位，… がある
- 10進数の "119" という数字が 119 という数を表すのは，1 の位が 9，10の位が 1，100 の位が 1 ということから $1 \cdot 9 + 10 \cdot 1 + 100 \cdot 1 = 9 + 10 + 100 = 119$ という計算があるから．
- では，2進数の "10101" という数字はどんな数を表すのだろう？
    1. この2進数は 1の位が1，2の位が0，4の位が1，8の位が0，16の位が1
    1. このことを10進数の数字を数として計算した例を参考に: $1\cdot1 + 0\cdot2 + 1\cdot4 + 0\cdot8 + 1\cdot16 = 1 + 4 + 16 = 21$

**Q2** 逆に10進数の21は2進数の10101で表されるはずだけれど，それはどうやって理解すればいいんだろう？

- 321 という数が十進数の数字 "321" で表されることについて考えてみよう．

    - まず，321 を十進数の底の 10 で割ると 32 余り 1．この「余り 1」を数字の "1" として覚えておく．
    - つぎに，商の 32 を 10 で割ると 3 余り 2．この「余り 2」を数字の "2" として覚えておく．
    - さらに，商の 3 を 10 で割ると 0 余り 3．この「余り 3」を数字の "3" として覚えておく．

    - 商が 0 でやることがなくなったので，10 での割り算はおしまい．
    
    - 覚えておいた数字 "3", "2", "1" をつないで "321" が得られる．

- つぎに 21 という十進数を2進数で表すことを考える．今度は2進数なので，底は 2．

    - まず，21 を 2 で割ると 10 余り 1．この「余り 1」を数字の "1" として覚えておく．

    - つぎに，商の 10 を 2 で割ると 5 余り 0．この「余り 0」を数字の "0" として覚えておく．

    - さらに，商の 5 を 2 で割ると 2 余り 1．この「余り 1」を数字の "1" として覚えておく．

    - もう先が見えた！商の 2 を 2 で割ると 1 余り 0．この「余り 0」を数字の "0" として覚えておく．

    - 最後に，商の 1 を 2 で割ると 0 余り 1．この「余り 1」を数字の "1" として覚えておく．

    - 商が 0 でやることがなくなったので，割り算はおしまい．

    - 覚えておいて数字をつないで "10101" が得られた．元の 2 進数と一致している．やった〜！

## データは数である → 3. 数で音を表現すること

音は時間とともに変化する空気の振動です．ある一瞬の音は音波によって引き起こされる周囲からの圧力のずれとして理解されています．これを**音圧**と呼ぶそうです． ← 詳しいことは物理の教員に訊ねて下さい．

音圧の時間変化を一定時間ごと，一定音圧ごとに量子化することで音を数値化できます．

![音のデータのコンピュータ内での表現](https://i.gyazo.com/a4f198ba148a0db9efaeebe9ae786285.png)

上の図では変化を曲線で表現され，それを時間方向，空気密度方向に量子化しています．曲線が通っている図 (c) の格子点を黒，そうでない格子点を白で表せば，「数で画像を表現する」方法を用いてこの音を表現することができます．

おそらく時間とともに変化する空気密度の値を数値列として表現した方がもっと小さなデータで表現することができるでしょう．

図 (c) で縦軸が対数表現になっていることに気づきましたか？これも小さなデータで表現する工夫です．

複雑なマルチメディアデータを小さなデータで扱うことは，実用的に重要です．たとえばスマホへの音楽データのダウンロードが短時間になったり，それに必要なパケット数を少なくする効果があります．このようにデータの表現をより小さな別の表現で置き換える技術を**データ圧縮**と呼びます．

## データは数である → 4. 数で映像を表現すること

テレビの画面は1秒間に30回以上書き換えています．滑らかに映っているように見えますが，実は超高速紙芝居を見せている機械と見做すことができます．

```python
x = int(input('自然数 x = '))
y = int(input('自然数 y = '))
積 = 0
while y > 0:
    積 = 積 + x
    y = y - 1
print(積)
```

乗算を $\pm1$ と繰り返しだけで実現しよう．以下のプログラムは上のプログラムと同じ内容です．このプログラムでは加算を使っているので題意を満たしません．

## 止まらない計算に注意

プログラムの作成に失敗するとプログラムの実行が止まらないことがあります．

たとえば，加算を計算したプログラムを少し変更した以下のプログラムを実行するとプログラムは止まらなくなります．

```python
a = int(input('自然数 a = '))
b = int(input('自然数 b = '))
和 = a
while b > 0:
    和 = 和 + 1
    b = b + 1
print(和)
```

このプログラムを実行して a, b をともに 1, 1 にするとこのプログラムは止まりません．プログラムの実行状態は以下の図のようになります．

> ![止まらないプログラムの実行](https://i.gyazo.com/5a80a8cdbf2befa1dfc4d71e27c99432.png)

画面の左上にある丸い印の縁を円弧がグルグルと回転しますが，これはプログラムが実行中であることを示しています．プログラムの行番号の左に表示されている小さな緑色の三角は，実行のある時点でこの行を実行したことを示しています．

このままでは埒があきません．実行を強制的に止めるには，グルグルと円弧が回っている丸をクリックして下さい．強制停止に伴なうエラーメッセージが表示されますけれど，計算を止めることができます．

計算をずっと放置すると，いずれ Google が怒って勝手に計算を止めることがあります．そこまでやってしまうとしばらく Google Colaboratory を利用できなくなる場合があります．そうなる前に自分で停めましょう．意味のない計算にコンピュータを浪費するのは環境にもよくありません．

## 例3: 囚われのスマイルくん

```python
import time

d1 =  1000000000000000000000000000
d2 =  1000000000110000110000000000
d3 =  1000000000110000110000000000
d4 =  1000000000000000000000000000
d5 =  1000001100000000000011000000
d6 =  1000000110000000000110000000
d7 =  1000000011000000001100000000
d8 =  1000000000111111110000000000
d9 =  1000000000000000000000000000
d10 = 1000000000000000000000000000

t = 0
n = 1   # 1, 1000, 1000000, 1000000000 と変化する

while t < 10:
    print(d1)
    print(d2)
    print(d3)
    print(d4)
    print(d5)
    print(d6)
    print(d7)
    print(d8)
    print(d9)
    print(d10)
    print()                      # 空行を出力

    time.sleep(1)                  # 1秒休む

    n10 = n * 10

    a   = d1 // n10
    b   = d1 - d1 // n * n
    d1  = (a * 10 + 1) * n + b
    a   = d2 // n10
    b   = d2 - d2 // n * n
    d2  = (a * 10 + 1) * n + b
    a   = d3 // n10
    b   = d3 - d3 // n * n
    d3  = (a * 10 + 1) * n + b
    a   = d4 // n10
    b   = d4 - d4 // n * n
    d4  = (a * 10 + 1) * n + b
    a   = d5 // n10
    b   = d5 - d5 // n * n
    d5  = (a * 10 + 1) * n + b
    a   = d6 // n10
    b   = d6 - d6 // n * n
    d6  = (a * 10 + 1) * n + b
    a   = d7 // n10
    b   = d7 - d7 // n * n
    d7  = (a * 10 + 1) * n + b
    a   = d8 // n10
    b   = d8 - d8 // n * n
    d8  = (a * 10 + 1) * n + b
    a   = d9 // n10
    b   = d9 - d9 // n * n
    d9  = (a * 10 + 1) * n + b
    a   = d10 // n10
    b   = d10 - d10 // n * n
    d10 = (a * 10 + 1) * n + b

    n = n * 1000
    t = t + 1
```
