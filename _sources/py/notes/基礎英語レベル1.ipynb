# 中学生の基礎英語レベル１

## ごあいさつ

「中学生の基礎英語シリーズ」は中学校で学習する英文法を２年間で学ぶシリーズ

本多先生のレベル１は「楽しく・ムリなく」力が身についていく仕掛けに満ちている。

**ストーリーを楽しもう**
- レッスンの核となるのがストーリー：テーマはファンタジー小説が大好きな中学生と、おしゃべりができるカメラのちょっと不思議な友情
- その日に学習する文法、すでに学習した内容が自然な会話に盛り込まれている。

**使える英語力をみがこう**
英語で対話できる力を養うために重要なのが**文法**。
- つまずきやすいところを重点的に解説。
- 豊富な練習問題
- レッスンの仕上げはアメリカ出身の2人による実践演習。インプットした知識をアウトプット。

身につけるコツは
- 自分から楽しんで学ぶ
- 少しずつでも触れる
- 毎日15分の放送

## 本多敏幸先生からのメッセージ

- すべてを詳細に理解しなくてもいい
- 正確に話せなくてもいい
- 音を大事にする
- 15分の放送 + あなた独自の学習

# 年間計画表

1. できること・でいないことについて言える / 名前・年齢・出身地について言える

    be動詞、一般動詞、助動詞 (can)
1. 過去のできごとについて話すことができる / 警告したり、注意したりできる

    過去形、命令文、動名詞
1. 自分や相手、家族や友だちについて話すことができる / 相手の持ち物についてたずねたり、答えたりできる

    ３単現在のs、人称代名詞、指示代名詞
1. 計画や予定などについて話すことができる / 義務・必要について言える

    未来表現、助動詞 (must)
1. 「夏季復習シリーズ」

1. 進行中の活動について話すことができる / 許可を求めたり、与えたりできる

    現在進行形、助動詞 (may, should)
1. 人や物の外見や特徴について言える / 自分の住んでいる場所・地域・国について話すことができる

    SVC (look や taste など)、There is / are 〜 の文
1. 身の周りで起こるできごとについて話すことができる / 自分や相手、家族や友だちについてたずねたり、答えたりできる

    接続詞 (because, when)、過去進行形
1. 感想や考えを言える / 知っているか、知らないかたずねたり、答えたりできる

    接続詞 (that)、不定詞（名詞的用法）
1. 提案やアドバイスを求めたり、与えたりできる / 話題に関連する質問をすることができる

    SVOO (make や give など)、SVOC (call や name など)
1. 自分の言うことを例や理由をあげて具体的に話すことができる / 賛成・反対についてたずねたり、答えたりできる

    不定詞（副詞的用法）、前置詞句による後置修飾
1. これまでの復習

## テキストの構成

### 月曜〜木曜はストーリーを聞く
- Get the **Idea!**: おおまかな内容の理解
- Get the **Point!**: 聞き取りのポイント （巻末に解答）
- Today's **Story**: ストーリーの内容の確認
- **Words** and **Phrases**: 単語や表現
- **Today's Key Point!**: 特に重要な文法項目、図解
    - Try **Repeating!**: 例文の繰り返し練習
    - **Say* it!: アメリカ人パートナと会話する
    - Be a **Performer!**: ストーリーのキャラクターを演ずる

### 金曜

- 今週の復習
    - Questions about **the Story**: Q/A
    - More Words and Phrases: 語句の深掘り
    - Key Points of the Week: 応用練習
    - Friday Special: 週替わりでおさえておきたいところ
- 付録: キーフレーズ・リストカード（単語帳？）

## 番組の構成

1. ストーリ x 2
    1. 全体を把握
    1. 質問の答えの部分に集中
1. 英文の意味を確認: 2つくらいの語句や表現の解説; もう一度、ストーリーを聞く
1. キーフレーズ（文法項目）
1. キーフレーズの練習
1. パートナーたちと対話
1. テキストを見ないで役割を演じる

## 復習のしかた

1. ストーリーの音読: つっかえているところの練習 / 音声教材
1. 単語の復習: テキスト巻末の重要単語
1. キーフレーズの復習: 巻末や付録のカード
