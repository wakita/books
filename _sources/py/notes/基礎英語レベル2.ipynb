# CAN-DO 年間計画表

1. １日の生活や日課について話すことができる / 何かを頼んだり、人の以来を受けたり断ったりできる

    助動詞（can, could）、SVOO、動名詞、頻度を表す副詞

1. 人やものの外見や特徴について言える / 相手を誘ったり、人の誘いに応じたりできる

    助動詞（may, will, shall）、接続詞（that）、前置詞句による後置修飾

1. 義務・必要について言える / 提案やアドバイスを求めたり、与えたりできる

    助動詞 (must, should)、have to do、不定詞（名詞的用法・副詞的用法）

1. 過去の経験やできごとについて話すことができる / 数や量についてたずねたり、答えたりできる

    不定詞（形容詞的用法）、接続詞（because, when, if）、SVOC、数の表現

1. 復習

1. ものや人を比較して説明したり、意見を言ったりできる / 自分や相手、家族や友だちなどについて話すことができる

    比較級（形容詞）、最上級（形容詞）、同等表現、否定表現

1. 感想や考えを言える / これまでのできごとや経験について話すことができる

    現在完了形、It is ... for 〜 to do

1. メニューなどを見て、食べ物や飲み物についてたずねたり、答えたりできる / 学校や季節の行事について話すことができる

    現在分詞、過去分詞、受け身、句動詞

1. 自分の持ち物について言える / 人やものの外見や特徴についてたずねたり、答えたりできる

    関係代名詞、比較級（副詞）、最上級（副詞）、some/any

1. 身の周りで起こるできごとについて話すことができる / 心配する気持ち・思いやりの気持ちを表現できる

    間接話法、間接疑問文、SVO + to do、代動詞

1. 賛成・反対について自分の考えを言える / 感謝の気持ちを表したり、受け取ったりできる

    現在完了進行形、部分否定

1. 不満・失望・後悔などを表現できる / 夢や希望について言える

    仮定法過去、SVO + do
